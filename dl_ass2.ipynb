```python
# McCulloch-Pitts Neuron for AND Gate

def mp_and_gate(x1, x2):
    # Weights
    w1 = 1
    w2 = 1

    # Threshold
    theta = 2

    # Weighted sum
    weighted_sum = w1 * x1 + w2 * x2

    # Activation function (step function)
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the AND gate
inputs = [(0,0), (0,1), (1,0), (1,1)]

print("x1 x2 -> Output")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_and_gate(x1, x2))
```

```python
# McCulloch-Pitts Neuron for OR Gate

def mp_or_gate(x1, x2):
    # Weights
    w1 = 1
    w2 = 1

    # Threshold
    theta = 1

    # Weighted sum
    weighted_sum = w1 * x1 + w2 * x2

    # Step activation function
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the OR gate
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("x1 x2 -> Output")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_or_gate(x1, x2))
```

```python
# McCulloch-Pitts Neuron for NAND Gate

def mp_nand_gate(x1, x2):
    # Weights
    w1 = -1
    w2 = -1

    # Threshold
    theta = -1

    # Weighted sum
    weighted_sum = w1 * x1 + w2 * x2

    # Step activation function
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the NAND gate
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("x1 x2 -> Output")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_nand_gate(x1, x2))
```

```python
# McCulloch-Pitts Neuron for NOR Gate

def mp_nor_gate(x1, x2):
    # Weights
    w1 = -1
    w2 = -1

    # Threshold
    theta = 0

    # Weighted sum
    weighted_sum = w1 * x1 + w2 * x2

    # Step activation function
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the NOR gate
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("x1 x2 -> Output")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_nor_gate(x1, x2))
```

```python
# McCulloch-Pitts Neuron for NOT Gate

def mp_not_gate(x):
    # Weight
    w = -1

    # Threshold
    theta = 0

    # Weighted sum
    weighted_sum = w * x

    # Step activation function
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the NOT gate
inputs = [0, 1]

print("x -> Output")
for x in inputs:
    print(x, "->", mp_not_gate(x))
```

```python
# McCulloch-Pitts Neuron for AND-NOT Gate (x1 AND NOT x2)

def mp_and_not_gate(x1, x2):
    # Weights
    w1 = 1     # excitatory
    w2 = -1    # inhibitory

    # Threshold
    theta = 1

    # Weighted sum
    weighted_sum = w1 * x1 + w2 * x2

    # Step activation function
    if weighted_sum >= theta:
        return 1
    else:
        return 0


# Testing the AND-NOT gate
inputs = [(0,0), (0,1), (1,0), (1,1)]

print("x1 x2 -> Output")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_and_not_gate(x1, x2))
```

```python
class MPNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def activate(self, inputs):
        net_input = sum(w * x for w, x in zip(self.weights, inputs))
        return 1 if net_input >= self.threshold else 0


# Given data
w1 = 1
w2 = 1
theta = 2
x1 = 1
x2 = 0

# Create MP neuron
mp_neuron = MPNeuron(weights=[w1, w2], threshold=theta)

# Compute output
output = mp_neuron.activate([x1, x2])

# Display result
print("Given values:")
print(f"x1 = {x1}, x2 = {x2}")
print(f"w1 = {w1}, w2 = {w2}")
print(f"Threshold (θ) = {theta}")

print("\nNet input =", w1*x1 + w2*x2)
print("AND gate output =", output)
```

## using tensor flow and mp neurons

```python
import tensorflow as tf

# Inputs for AND gate
X = tf.constant([[0., 0.],
                 [0., 1.],
                 [1., 0.],
                 [1., 1.]])

# Weights and threshold (MP neuron)
W = tf.constant([[1.],
                 [1.]])
theta = 2.0

# Weighted sum
net = tf.matmul(X, W)

# Step activation function
output = tf.cast(net >= theta, tf.int32)

print("Inputs:\n", X.numpy())
print("AND Gate Output:\n", output.numpy())
```
