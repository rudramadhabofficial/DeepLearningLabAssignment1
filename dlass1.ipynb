## Scalar and Vector Multiplication

```python
import tensorflow as tf

# Scalar
scalar = tf.constant(5, dtype=tf.float32) # Changed to float32

# Vector
vector = tf.constant([1, 2, 3], dtype=tf.float32)

# Scalar-Vector Multiplication
result = scalar * vector

print("Vector:", vector.numpy())
print("Scalar:", scalar.numpy())
print("Scalar * Vector:", result.numpy())
```

# 2. Matrix Addition, Subtraction, and Multiplication

```python
import tensorflow as tf

# Define two matrices
A = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
B = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

# Matrix Addition
addition = tf.add(A, B)

# Matrix Subtraction
subtraction = tf.subtract(A, B)

# Matrix Multiplication
multiplication = tf.matmul(A, B)

print("Matrix A:\n", A.numpy())
print("Matrix B:\n", B.numpy())
print("A + B:\n", addition.numpy())
print("A - B:\n", subtraction.numpy())
print("A * B:\n", multiplication.numpy())
```

## 3. Matrix Inverse and Transpose

```python
import tensorflow as tf

# Define a square matrix
A = tf.constant([[4, 7], [2, 6]], dtype=tf.float32)

# Transpose
transpose = tf.transpose(A)

# Inverse
inverse = tf.linalg.inv(A)

print("Matrix A:\n", A.numpy())
print("Transpose of A:\n", transpose.numpy())
print("Inverse of A:\n", inverse.numpy())
```

## 4. Eigenvalues and Eigenvectors

```python
import tensorflow as tf

# Define a square matrix
A = tf.constant([[2, 1], [1, 2]], dtype=tf.float32)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = tf.linalg.eig(A)

print("Matrix A:\n", A.numpy())
print("Eigenvalues:\n", eigenvalues.numpy())
print("Eigenvectors:\n", eigenvectors.numpy())
```

## 5. Check Orthogonality of a Matrix

```python
import tensorflow as tf

# Define an orthogonal matrix
A = tf.constant([[0, -1], [1, 0]], dtype=tf.float32)

# Compute A^T * A
product = tf.matmul(tf.transpose(A), A)

# Identity matrix
identity = tf.eye(A.shape[0])

# Check orthogonality
is_orthogonal = tf.reduce_all(tf.abs(product - identity) < 1e-5)

print("Matrix A:\n", A.numpy())
print("A^T * A:\n", product.numpy())
print("Is Orthogonal:", is_orthogonal.numpy())
```
